## Zadanie 4

Cel: użycie warunku i pętli.

Instrukcja:

1. Ustaw: `punkty = [45, 67, 82, 39, 91]`.
2. Dla każdego wyniku wpisz do listy `status`:
   - `zaliczone` jeśli wynik >= 50,
   - `niezaliczone` w przeciwnym razie.
3. Policz, ile osób zaliczyło.

Wymagane konstrukcje:

- `for`
- `if/else`
- `list.append()`
- `list.count()`

Kryterium zaliczenia:

- lista `status` ma tyle elementów co `punkty`,
- liczba `zaliczone` jest poprawna.

Checkpoint:

- pokaż listę `status` i liczbę zaliczeń.


```python
# Miejsce na Twoje rozwiązanie.
punkty = [45, 67, 82, 39, 91]
status = []
for wynik in punkty:
    if wynik >= 50:
        status.append("zaliczone")
    else:
        status.append("niezaliczone")

print(status)
print(status.count("zaliczone"))
x = ["zaliczone" if wynik >= 50 else "niezaliczone" for wynik in punkty]
x
```

## Zadanie 5

Cel: napisanie pierwszej funkcji.

Instrukcja:

1. Zdefiniuj funkcję `policz_procent(punkty, maks=100)`.
2. Funkcja ma zwracać procent: `(punkty / maks) * 100`, zaokrąglony do 2 miejsc.
3. Uruchom funkcję dla `37`, `88`, `100`.

Wymagane funkcje:

- `def`
- `return`
- `round()`

Kryterium zaliczenia:

- funkcja zwraca poprawne wartości procentowe.

Checkpoint:

- pokaż definicję i trzy wyniki.

Dokumentacja:

- https://docs.python.org/3/tutorial/controlflow.html#defining-functions


```python
# Miejsce na Twoje rozwiązanie.
def policz_procent(punkty, maks=100):
    return round((punkty / maks) * 100, 2)
print(policz_procent(37, 50))
print(policz_procent(88, 100))
print(policz_procent(100, 150))
```
